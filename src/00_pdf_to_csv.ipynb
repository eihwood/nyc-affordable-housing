```python
# Queens Prep
outnames = files[2].split('-')[3]
fileout = "../data/L1/2020-DHCR-Bldg-" + outnames + "_geocoded.csv"
print(fileout)

user_agent = "example app"
data = pd.read_csv(files[2])
    


# BLDGNO1 sometimes has multiple building numbers associated with same building... to avoid replication of same 'address' only take first number
data['BLDGNO1_New'] = data['BLDGNO1'].str.split(' ').str[0]

# CITY and COUNTY # ARE COMBINED FOR QUEENS - LETS SEPARATE THEM
data['CITY'] = data['CITY COUNTY'].str.replace('\d+', '')

# add series 'Address' which we will use to geolocate
data['Address'] = data['BLDGNO1_New'] + ' ' + data['STREET1'] + ' ' + data['STSUFX1'] + ' ' + data['CITY'] + ' NY ' + data['ZIP']
```

```python
# Apply geocode function to queens dataset

geocode_func(data=data, fileout=fileout,address_col_name="Address")
```
